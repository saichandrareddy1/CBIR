```python
from scipy.ndimage import imread
images = []
for i in range(len(data)):
    img = imread(f"{data[i]}")
    img = cv2.resize(img, (48, 48))
    #print(img.shape)
    img = img.flatten().reshape(1, -1)[0].tolist()
    #print(img)
    images.append(img)
```
